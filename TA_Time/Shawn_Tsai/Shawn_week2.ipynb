```python
import pandas as pd
df=pd.read_csv('week2.csv')
df.head()
```

```python
#減少欄位這樣show出來比較好理解
#最有可能是被第0列的中文字給干擾
#所以把第0列drop掉，得到temp_new
temp=df[['site_id','edu_age_15up_total']]
temp_new = temp.drop(0)
print(temp.head())
print(temp_new.head())
print(temp.info())

#把只剩數字的edu_age_15up_total轉成int
temp_new['edu_age_15up_total']=temp_new["edu_age_15up_total"].astype(int)
print(temp_new.head())
print(temp_new.info())
```

```python
#補充說明
#如果沒有drop掉，他會很困擾文字怎麼轉成int
temp['edu_age_15up_total']=temp["edu_age_15up_total"].astype(int)
```
